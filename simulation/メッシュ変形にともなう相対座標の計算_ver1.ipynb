```python
import numpy as np
import matplotlib.pylab as plt
```

```python
class Vector2:
    def __init__(self, x=0., y=0.):
        self.x = x
        self.y = y
        
    def set(self, x, y):
        self.x = x
        self.y = y
```

```python
#メッシュの定義
b = Vector2(-0.4, 0.4)
d = Vector2( 0.0, 0.4)
a = Vector2(-0.4, 0.0)
c = Vector2( 0.0, 0.0)

#三角形：ABC, DCB
```

```python
def drawMesh(fig, a, b, c, d, color="black"):
    #ABC
    fig.plot([a.x, b.x], [a.y, b.y], color=color)
    fig.plot([b.x, c.x], [b.y, c.y], color=color)
    fig.plot([c.x, a.x], [c.y, a.y], color=color)
    fig.text(a.x*1.04, a.y*1.04, "a", horizontalalignment='center', verticalalignment='center', color=color)
    fig.text(b.x*1.04, b.y*1.04, "b", horizontalalignment='center', verticalalignment='center', color=color)
    fig.text(c.x*1.04, c.y*1.04, "c", horizontalalignment='center', verticalalignment='center', color=color)
    #DCB
    fig.plot([d.x, c.x], [d.y, c.y], color=color)
    #g.plot([c.x, b.x], [c.y, b.y], color=color)
    fig.plot([b.x, d.x], [b.y, d.y], color=color)
    #fig.plot(p.x, p.y, "o", color=color)
    fig.text(d.x*1.04, d.y*1.04, "d", horizontalalignment='center', verticalalignment='center')


    fig.set_xlim(-0.5, 0.1)
    fig.set_ylim(-0.1, 0.5)
```

```python
def calcParams(a, b, c, d, p):
    ad = Vector2(d.x-a.x, d.y-a.y)
    bc = Vector2(c.x-b.x, c.y-b.y)
    ap = Vector2(p.x-a.x, p.y-a.y)

    s = (ap.x*bc.y - bc.x*ap.y) / (ad.x*bc.y - bc.x*ad.y)
    t = (ad.x*ap.y - ap.x*ad.y) / (ad.x*bc.y - bc.x*ad.y)
    return s, t
```

```python
def calcPosition(a, b, c, d, s, t):
    ad = Vector2(d.x-a.x, d.y-a.y)
    bc = Vector2(c.x-b.x, c.y-b.y)

    p_new = Vector2(s*ad.x + t*bc.x + a.x, s*ad.y + t*bc.y + a.y)
    return p_new
```

```python
#新しいメッシュ
b2 = Vector2(-0.2, 0.2)
d2 = Vector2( 0.0, 0.2)
a2 = Vector2(-0.2, 0.0)
c2 = Vector2( 0.1, 0.1)

p2 = calcPosition(a2, b2, c2, d2, s, t)

fig, ax = plt.subplots(figsize=(10, 10))
drawMesh(ax, a, b, c, d, color="black")
drawMesh(ax, a2, b2, c2, d2, color="red")
ax.plot(p.x, p.y, "o", color="black")
ax.plot(p2.x, p2.y, "o", color="red")
ax.set_aspect("equal")
plt.show()
```

```python
#円
thetas = np.linspace(0, 2*np.pi, 100)
ps_circle = []
for theta in thetas:
    ps_circle.append(Point(np.cos(theta)*0.2 - 0.2, np.sin(theta)*0.2 + 0.2))

#新しいメッシュ
a2 = Vector2(-0.2, 0.2)
b2 = Vector2( 0.0, 0.2)
c2 = Vector2(-0.2, 0.0)
d2 = Vector2( 0.1, 0.1)

fig, ax = plt.subplots(figsize=(10, 10))
drawMesh(ax, a, b, c, d, color="black")
drawMesh(ax, a2, b2, c2, d2, color="red")

for p_circle in ps_circle:
    s, t = calcParams(a, b, c, d, p_circle)
    p_new = calcPosition(a2, b2, c2, d2, s, t)

    ax.plot(p_circle.x, p_circle.y, "o", color="black")
    ax.plot(p_new.x, p_new.y, "o", color="red")
ax.set_aspect("equal")
plt.show()
```
